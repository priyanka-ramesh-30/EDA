# VIDEO GAMES SALES DATASET
ABOUT THE DATASET:
This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

Fields include

*   Rank - Ranking of overall sales
*   Name - The games name
*   Platform - Platform of the games release (i.e. PC,PS4, etc.)
*   Year - Year of the game's release
*   Genre - Genre of the game
*   Publisher - Publisher of the game
*   NA_Sales - Sales in North America (in millions)
*   EU_Sales - Sales in Europe (in millions)
*   JP_Sales - Sales in Japan (in millions)
*   Other_Sales - Sales in the rest of the world (in millions)
*   Global_Sales - Total worldwide sales (in millions)

```python
from numpy import mean
from numpy import std
data_mean, data_std = mean(new_df['JP_Sales']), std(new_df['JP_Sales'])
print(data_mean)
print(data_std)
cut_off = data_std * 3
print(cut_off)
lower, upper = data_mean - cut_off, data_mean + cut_off
outliers = [x for x in new_df['JP_Sales'] if x < lower or x > upper]
print(outliers)
print('Identified outliers: %d' % len(outliers))
```

```python
print(new_df['JP_Sales'].quantile(0.50))
print(new_df['JP_Sales'].quantile(0.95))
print(new_df['JP_Sales'].quantile(0.05))
print(new_df['JP_Sales'].quantile(0.25))
new_df['JP_Sales']=np.where(new_df['JP_Sales']>0.16,0.0,new_df['JP_Sales'])
```
